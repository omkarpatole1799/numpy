```python
# vertical stacking
m1 = np.array([
  [1,2,3],
  [4,5,6]
])

m2 = np.array([
  [3,2,1],
  [6,5,4]
])

print(m1)
print(m2)

vstacked  = np.vstack([m1,m2])
print(vstacked)
```

```python
# horizontal stacking

hstacked = np.hstack((m1, m2))
print(hstacked)
```

```python

# read file 
file =(np.genfromtxt("data.txt", delimiter=","))
print(file)
```

```python
print(file.astype("int32"))
```

```python
# boolen masking and advanced indexing
f = file <= 7
print(f)

e = file[file <= 9]
print(e)
```

```python
# get elemnet by index

print(e[[1]])
print(e[[5,6]])
```

```python
print(np.all(f >= 2))
```
