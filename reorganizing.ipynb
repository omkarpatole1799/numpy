```python
import numpy as np
```

```python
before = np.array([[1, 2, 3], [7, 8, 9]])
```

```python
print(before)
print(before.shape)
```

```python
after = before.reshape((3,2))
print(after)
print(after.shape)
```

```python
# vertical stacking
m1 = np.array([
  [1,2,3],
  [4,5,6]
])

m2 = np.array([
  [3,2,1],
  [6,5,4]
])

print(m1)
print(m2)

vstacked  = np.vstack([m1,m2])
print(vstacked)
```

```python
# horizontal stacking

hstacked = np.hstack((m1, m2))
print(hstacked)
```
