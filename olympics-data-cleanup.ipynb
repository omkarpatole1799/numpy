```python
import pandas as pd
```

```python
df_olympics = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/Olympics-Dataset/refs/heads/master/athletes/bios.csv')
```

```python
df_olympics.sample(3)
df = df_olympics.copy()
```

```python
# cleanup full name remove the '•' string
df["name"] = df['Full name'].str.replace('•',' ')
# df[["name", "Full name"]].sample(5)
```

```python
# split measurements into height_cm and weight_kg

# df.sample()

df[['height_cm', 'weight_kg']] = df["Measurements"].str.split('/', expand=True)
```
