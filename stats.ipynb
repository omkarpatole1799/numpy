```python
import numpy as np
```

```python
# find min

stats = np.array([
  [1,2,3],
  [4,5,6],
  [8,8,9]
])

# print(stats.shape)
# print(stats)
print(np.min(stats, axis=1))
```

```python
#print max
print(np.max(stats, axis =1))
print(np.max(stats, axis =0))
```

```python
# sum
print(np.sum(stats))
print(np.sum(stats, axis = 1))
print(np.sum(stats, axis = 0))
```
