```python
olympics = pd.read_csv("data/bios.csv")
print(olympics)
```

```python
# get atheletes who has height greater than 200 cm
# oly_h_d = olympics.loc[(olympics.height_cm > 200), ["name","born_date","height_cm"]].sort_values(by="height_cm")
# print(oly_h_d)

print(len(olympics))

olympics_bio_filterd_by_height = olympics.loc[(olympics["height_cm"] >=190), ["name", "born_date", "height_cm"]]
sorted_by_height = olympics_bio_filterd_by_height.sort_values(by="height_cm", ascending=False)
print(olympics_bio_filterd_by_height)
print(len(sorted_by_height))
```

```python
# we can add multiple conditions in loc for filtering data
# e.g. get player names whose height is above 200 cm and born country = USA

# print(olympics)

filtered_data = olympics.loc[(olympics["height_cm"] >= 190) & (olympics["born_country"] == "USA"), ["name", "born_country", "height_cm"]]
print(filtered_data)
```

```python
import pandas as pd
import numpy as np
olybios = pd.read_csv("data/bios.csv")
nocs = pd.read_csv("data/noc_regions.csv")
print(olybios.head())
print(nocs.head())
```

```python
# merging two data frames / joining two data frames
# this is similar to mysql joins
# pd.merge(df1, df2, left_on="", right_on="", how="left | inner | right")

merged_new = pd.merge(olybios, nocs, left_on="born_country", right_on="NOC", how="left")
merged_new.rename(columns={"region": "full_country_name"}, inplace=True)
merged_new.head()
```

```python
print(merged_new[["name","full_country_name","born_country"]])
```

```python
# concat two df

fra_df = merged_new[merged_new["born_country"] == "FRA"]
gbr_df = merged_new[merged_new["born_country"] == "GBR"]


# print(fra_df.head())
# print(gbr_df.head())

# concat fra_df and gbr_df
new_merged = pd.concat([fra_df,gbr_df])
print(new_merged.tail()[["name","born_country"]])
```

```python
df_coffee = pd.read_csv("warmup-data/coffee.csv")
df_coffee
```

```python
# groupby function for aggregation of data
# get total units sold by each coffee type in a given data
df_coffee.groupby("Coffee Type")["Units Sold"].sum()

```

```python
# get average coffee units sold for each coffee type

print(df_coffee.groupby("Coffee Type")["Units Sold"].mean())
```

```python
# add price column in df_coffee
df_coffee["Price"] = np.where(df_coffee['Coffee Type'] == 'Expresso',3.99, 6.99)
print(df_coffee
      )

# add a column total revenue
df_coffee["Total Revenue"] = df_coffee["Units Sold"] * df_coffee["Price"]

print(df_coffee
      )
```

```python
# aggreggate by multiple columns
# .agg function
df_coffee.groupby('Coffee Type').agg({
  "Units Sold": "sum",
  "Total Revenue": "mean"
})
```

```python
# calcluate sum per day
df_coffee.groupby("Day").agg({
  "Units Sold": "sum"
}).sort_values(by="Units Sold")
```
