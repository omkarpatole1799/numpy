```python
import numpy as np
```

```python
# multiply two matrices


a = np.ones((5,4)
            )
print(a)

b = np.full((4,2), 3)

print(b)

print(np.matmul(a,b))
```

```python
# find determinant of a matrix
c = np.array([[1, 4],[4,2]])
det = np.linalg.det(c)
print(det)
```

```python
d= np.ones([5,5])
print(d)
print(np.linalg.det(d))
```

```python
e = np.identity(5)
print(e)
print(np.linalg.det(e))
```
